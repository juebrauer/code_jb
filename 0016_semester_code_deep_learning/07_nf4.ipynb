```python
import numpy as np
import matplotlib.pyplot as plt

# Generate normal weights
weights = np.random.normal(0, 3, 50000)
#weights = np.random.uniform(0, 1, 10000)

# Approximate NF4 codebook using normal quantiles
# (This is for demonstration; real NF4 uses a hand‑crafted codebook.)
probsb = (np.arange(16) + 0.5) / 16
codebook = np.quantile(weights, probs)

# Plot histogram
plt.figure(figsize=(10,6))
plt.hist(weights, bins=80, density=True, color="orange")

# Add codebook lines
for v in codebook:
    plt.axvline(v, color="black", linestyle="--")

plt.title("Normalverteilte Gewichte & NF4‑ähnliches Codebook (Demo)")
plt.xlabel("Wert")
plt.ylabel("Dichte")
plt.tight_layout()
plt.show()
```
