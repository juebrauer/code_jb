# Alles in einem Skript

```python
import os
import tensorflow as tf
import pandas
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

# 1. Daten einlesen
URL = "http://www.juergenbrauer.org/datasets/0009_real_estates_ames.csv"
df = pandas.read_csv(URL)

# 2. Inputs und Outputs fÃ¼rs NN definieren
x = df[ ["YearBuilt", "GarageCars"] ]
y = df[ ["SalePrice"] ]

# 3. Beispieldaten --> Trainings- und Testdaten
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# 4. Skalierung der Daten
s_input = MinMaxScaler()
s_output = MinMaxScaler()
s_input.fit(x_train)
s_output.fit(y_train)
x_train_scaled = s_input.transform(x_train)
y_train_scaled = s_output.transform(y_train)

# 5. MLP erstellen und trainieren
model = tf.keras.Sequential(name="Hauspreispraediktor")
model.add(tf.keras.layers.InputLayer(shape=(2,)))
model.add(tf.keras.layers.Dense(10, activation="relu", name="hidden1"))
model.add(tf.keras.layers.Dense(1, name="output"))
myoptimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss="mse", optimizer=myoptimizer)
model.summary()
h = model.fit(x_train_scaled,y_train_scaled, epochs=100, verbose=0)
plt.plot(h.history["loss"])
plt.show()
```

```python
x_test_scaled = s_input.transform(x_test)
x_test_scaled[:5]
```

```python
preds_scaled = model.predict(x_test_scaled, verbose=0)
preds_dollar = s_output.inverse_transform(preds_scaled).flatten()
preds_dollar[:5]
```

```python
gt = y_test.values.flatten()
gt[:5]
```

```python
import numpy
#    MEAN  ABS   ERROR (MAE)
numpy.mean(abs(preds_dollar - gt))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(gt, preds_dollar)
```

# Das finale Skript

```python
import os
import tensorflow as tf
import pandas
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

# 1. Daten einlesen
URL = "http://www.juergenbrauer.org/datasets/0009_real_estates_ames.csv"
df = pandas.read_csv(URL)

# 2. Inputs und Outputs fÃ¼rs NN definieren
x = df[ ["YearBuilt", "GarageCars", "OverallQual", "Neighborhood"] ]
x = pandas.get_dummies(x, dtype=int)
y = df[ ["SalePrice"] ]

# 3. Beispieldaten --> Trainings- und Testdaten
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

# 4. Skalierung der Daten
scale_data = True
scale_type = "stdscaler"
if scale_data:
    if scale_type == "minmax":
        s_input = MinMaxScaler()
        s_output = MinMaxScaler()
    elif scale_type == "stdscaler":
        s_input = StandardScaler()
        s_output = StandardScaler()
    s_input.fit(x_train)
    s_output.fit(y_train)
    x_train_scaled = s_input.transform(x_train)
    y_train_scaled = s_output.transform(y_train)
else:
    x_train_scaled = x_train
    y_train_scaled = y_train

# 5. MLP erstellen und trainieren
model = tf.keras.Sequential(name="Hauspreispraediktor")
nr_inputs = x_train_scaled.shape[1]
model.add(tf.keras.layers.InputLayer(shape=(nr_inputs,)))
model.add(tf.keras.layers.Dense(20, activation="relu", name="hidden1", use_bias=True))
model.add(tf.keras.layers.Dense(1, name="output"))
myoptimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss="mse", optimizer=myoptimizer)
model.summary()
h = model.fit(x_train_scaled,y_train_scaled, epochs=300, verbose=0)
plt.plot(h.history["loss"])
plt.show()

# 6. Wie gut ist der HauspreisprÃ¤diktor?
if scale_data:
    x_test_scaled = s_input.transform(x_test)
    preds_scaled = model.predict(x_test_scaled, verbose=0)
    preds_dollar = s_output.inverse_transform(preds_scaled).flatten()
else:
    preds_dollar = model.predict(x_test, verbose=0)
gt = y_test.values.flatten()
MAPE = mean_absolute_percentage_error(gt, preds_dollar)*100
print(f"Im Mittel kÃ¶nnen wir den Verkaufspreis auf {MAPE:.1f} % genau schÃ¤tzen!")
```
