# Vergleichs-Programm in Python

```python
datei = open("zahlen.txt", mode="rt")
txt = datei.read()
datei.close()
tokens = txt.split()
summe = 0.0
for token in tokens:
    zahl = float(token)
    summe = summe + zahl
print(summe)
```

# Wie sieht OOP in Python aus?

```python
import this
```

```python
class Vector:

    def __init__(self, elements):
        self.elements = elements

    def __add__(self, other):
        result_vec = Vector( [self.elements[0]+other.elements[0],
                              self.elements[1]+other.elements[1],
                              self.elements[2]+other.elements[2]
                             ]
                            )
        return result_vec
```

```python
v1 = Vector( [10,20,30] )
v2 = Vector( [50,60,70] )
```

```python
v1.elements
```

```python
v2.elements
```

```python
[10,20,30] + [40,50,60]
```

```python
v3 = v1 + v2
```

```python
v3.elements
```
