# Aus einem Beispielvideo einzelne Beispielframes herausnehmen

```python
import cv2 # pip install opencv-python
import os

# Pfad zur MKV-Datei
video_path = "/home/juebrauer/link_to_vcd/10_datasets/60_brauer_dataset/brauer.mkv"

# Zielordner f√ºr die Frames
output_dir = "brauer_frames"
os.makedirs(output_dir, exist_ok=True)

cap = cv2.VideoCapture(video_path)

print(cap)

frame_index = 0
saved_index = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break  # Video fertig

    # Jedes 5. Frame speichern
    if frame_index % 5 == 0:
        filename = os.path.join(output_dir, f"frame_{saved_index:05d}.png")
        cv2.imwrite(filename, frame)
        saved_index += 1

    frame_index += 1

cap.release()
print("Fertig! Frames gespeichert in:", output_dir)
```
